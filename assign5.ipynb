```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
n_values = [2, 4, 6, 8]
y_values = [1/3, 1/5, 1/8, 1/10]

# Plot the values and connect them with a line
plt.plot(n_values, y_values, marker='o', linestyle='-')

plt.xlabel("Number of steps (n)")
plt.ylabel("Probability of going back to 1")

plt.title("Plot of Probabilities going back to 1 with different n values")

plt.show()
```

```python
n_values = [2, 4, 6, 8]  
y_values = [1/3, 1/5, 1/8, 1/10]  

# Define the exponential function to fit
def exp_func(x, a, b):
  return a * np.exp(-b * x)  # a: scaling factor, b: decay rate

# Fit the curve
popt, pcov = curve_fit(exp_func, n_values, y_values)

# Extract the fitted parameters
a = popt[0]  
b = popt[1]  

# Print the fitted parameters
print("Fitted parameters:")
print("  - Scaling factor (a):", a)
print("  - Decay rate (b):", b)

# Generate x-values for the fitted curve 
x_fit = np.linspace(min(n_values), max(n_values), 100)

# Calculate the fitted curve y-values
y_fit = exp_func(x_fit, a, b)


plt.plot(n_values, y_values, 'o',  linestyle='-', label='Actual probability')
plt.plot(x_fit, y_fit, '-', label='Fitted Curve')
plt.xlabel("Number of steps (n)")
plt.ylabel("Probability of going back to 1")
plt.title("Curve Fit")
plt.legend()
plt.show()
```
