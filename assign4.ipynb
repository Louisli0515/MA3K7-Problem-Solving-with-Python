```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def combination(n, k):
    if n > k:
        return math.factorial(n) / (math.factorial(k) * math.factorial(n - k))
    if n < k:
        return 0
```

```python
def probability_of_landing_on_number_within_n_tosses(n):
    probability = 0
    
    # Calculate the probability for each possible number of tosses leading to the target number
    for k in range((n - 1) // 2):
        probability += combination((n - 1) // 2 + k, (n - 1) // 2 - 1 - k) * (1/2) ** ((n - 1) // 2 + k)
    
    probability += (1/2) ** (n - 1)
    
    return probability

```

```python
n = 5 #Change n into various values to validate in the Review section
probability = probability_of_landing_on_number_within_n_tosses(n)
print(f"Probability of landing on number {n} within {n-1} tosses:", probability)
```

# Plot of even case

```python
def probability_of_n_t_within_even_n(n, n_t):
    probability = combination(n_t, n - n_t - 1) * (1/2) ** n_t
    return probability
```

```python
n = 100
n_t_values = list(range(n // 2, n))  # n_t ranges from n/2 to n-1
probabilities = [probability_of_n_t_within_even_n(n, n_t) for n_t in n_t_values]

plt.plot(n_t_values, probabilities, marker='o')
plt.xlabel("Value of n_t")
plt.ylabel("Probability")
plt.title(f"Probability of landing on each n_t value within {n-1} tosses")
plt.grid(True)
plt.show()
```

# Plot of odd case

```python
def probability_of_n_t_within_odd_n(n, n_t):
    probability = combination(n_t, n - n_t - 1) * (1/2) ** n_t
    return probability
```

```python
n = 99  # Change n to any desired odd number
n_t_values = list(range((n + 1) // 2, n))  # n_t ranges from (n-1)/2 to n-1
probabilities = [probability_of_n_t_within_odd_n(n, n_t) for n_t in n_t_values]

plt.plot(n_t_values, probabilities, marker='o')
plt.xlabel("Value of n_t")
plt.ylabel("Probability")
plt.title(f"Probability of landing on each n_t value within {n} tosses")
plt.grid(True)
plt.show()
```

# Chi-square Distribution

```python
df = 3  # Degrees of freedom
x = np.linspace(0, 20, 1000) 
y = np.exp(-x / 2) * (x / 2) ** (df / 2 - 1) / (2 ** (df / 2) * math.gamma(df / 2))

plt.figure(figsize=(8, 5))
plt.plot(x, y, label=f'Chi-square distribution (df={df})', color='blue')
plt.xlabel('x')
plt.ylabel('Probability density')
plt.title('Chi-square Distribution')
plt.legend()
plt.grid(True)
plt.show()
```
